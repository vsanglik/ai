```python
# 2 methods
pd.get_dummies(df['Sex'])
```

```python
len(df['Sex'])
```

```python
# df['Sex'].values - array
df['Sex'].values.reshape(len(df['Sex']),1)
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot = OneHotEncoder() # create object
onehot.fit(df['Sex'].values.reshape(len(df['Sex']),1) ) # apply fit() method --> Learn the unique values from that column
# Expects a 2d array
onehot.get_feature_names_out()
```

```python
result = onehot.transform(df['Sex'].values.reshape(len(df['Sex']),1))
result
```

```python
result.toarray() # convert the data to full matrix
```
